## Get Data Profile From Users

```python
data_FB_profile = []

profile = get_profile(name_user_account, cookies = {'sb': 'jYr1Y026A-2VjDm_XJBR6ETZ', 'datr': 'j4r1Y1Hi2f_ST4wVzobNFmCL', 'dpr': '1.5', 'c_user': '100001581864778', 'xs': '30%3AopPGizY1F1mghA%3A2%3A1677036201%3A-1%3A11175', 'fr': '0NJTocD2fx3WXJcHm.AWVFijqYbwIsqU_014egTaWO3ag.Bj9YqN.FZ.AAA.0.0.Bj9Yqt.AWVTp-49XAI', 'wd': '1280x569'})
    
#get employment data from user
try :
    start_year = '-'
    last_year = '-'
    company = '-'
    
    employment_data = profile['Pekerjaan'].split('\n')
    data_profile = employment_data[0:3]
    term = []

    for i in range(len(data_profile)):
        company = data_profile[0]
        term.append(re.findall(r'\d+', data_profile[i]))
    for j in range(len(term)):
        for k in range(len(term[j])):
            if term[j] == None:
                break
            else:
                if str(term[j][k]).isdigit() == True:
                    start_year = data_profile[j].split(' - ')[0]
                    last_year = data_profile[j].split(' - ')[-1]
                    if last_year == 'Sekarang':
                        last_year = 'Now'
                        #tahun_akhir = datetime.date.today()
                        #tahun_akhir = tahun_akhir.strftime("%d %B %Y")

except:
    start_year = '-'
    last_year = '-'
    company = '-'
        
# get general information from user
try:
    birth_day = '-'
    sex = '-'
    language = '-'
    
    general_info = profile['Info Umum'].split('\n')

    for i in range(1, len(general_info), 2):
        if general_info[i] == 'Tanggal Lahir':
            birth_day = general_info[i-1]
        elif general_info[i] == 'Jenis Kelamin':
            sex = general_info[i-1]
        else:
            language = general_info[i-1]

except:
    birth_day = '-'
    sex = '-'
    language = '-'
        
#get residence data from the user
try:
    hometown = '-'
    current_city = '-'
    
    place = profile['Pernah Tinggal di'].split('\n')
        
    for i in range(1, len(place), 2):
        if place[i] == 'Kota asal':
            hometown = place[i-1]
        elif place[i] == 'Kota Saat Ini':
            current_city = place[i-1]
except:
    hometown = '-'
    current_city = '-'
        
#get last education level data from user
try:
    last_education = profile['Pendidikan'].split('\n')
    school = last_education[0]
    level_education = last_education[1]
        
except:
    school = '-'
    level_education = '-'
        
#get family members data from user
try:
    family = profile['Anggota Keluarga'].split('\n')
    family_member_name = family[0]
    family_relationship = family[1]
    
except:
    family_member_name = '-'
    family_relationship = '-'
            
data_FB_profile.append({
        'Id_user': profile['id'],
        'Nama' : profile['Name'],
        'Tanggal_lahir' : birth_day,
        'Sex' : sex,
        'Asal' : hometown,
        'Domisili' : current_city,
        'Asal_Sekolah_Terakhir': school,
        'Pendidikan_Terakhir': level_education,
        'Tempat_Kerja(last)' : company,
        'Tahun_mulai' : start_year,
        'Tahun_akhir_bekerja' : last_year,
        'Nama_Anggota_Keluarga' : family_member_name,
        'Hubungan_Keluarga' : family_relationship})
```

## Data Postingan FB

```python
data_FB_post = []

postan = []
    
user = get_posts(name_user_account, pages=1, options={'reactors': True}, cookies= {'sb': 'jYr1Y026A-2VjDm_XJBR6ETZ', 'datr': 'j4r1Y1Hi2f_ST4wVzobNFmCL', 'dpr': '1.5', 'c_user': '100001581864778', 'xs': '30%3AopPGizY1F1mghA%3A2%3A1677036201%3A-1%3A11175', 'fr': '0NJTocD2fx3WXJcHm.AWVFijqYbwIsqU_014egTaWO3ag.Bj9YqN.FZ.AAA.0.0.Bj9Yqt.AWVTp-49XAI', 'wd': '1280x569'})
    
for post in user:
    postan.append(post)
        
    
for posting in postan:
        
    #get post_id
    id_post = posting['post_id']
        
    #get text from post
    try:
        text_postingan = posting['text']
            
        if text_postingan == '':
            text_postingan = 'Tidak ada data'
    except:
        text_postingan = '-'
        
    #get date posting
    time = posting['time'].strftime("%d %B %Y")
        
    #get reactions
    try:
        like = posting['reactions']['suka']
    except:
        like = 0
            
    try:
        care = posting['reactions']['care']
    except:
        care = 0
        
    try:
        laugh = posting['reactions']['haha']
    except:
        laugh = 0
            
    try:
        angry = posting['reactions']['angry']
    except:
        angry = 0
            
    try:
        sad = posting['reactions']['sad']
    except:
        sad = 0
            
    try:
        love = posting['reactions']['love']
    except:
        love = 0
            
    try:
        wow = posting['reactions']['wow']
    except:
        wow = 0
            
    data_FB_post.append({
            'Id_user': post['user_id'],
            'Nama' : post['username'],
            'post_id' : id_post,
            'waktu_upload': time,
            'teks_postingan' : text_postingan,
            'reaksi_like' : like,
            'reaksi_care' : care,
            'reaksi_laugh' : laugh,
            'reaksi_angry' : angry,
            'reaksi_sad' : sad,
            'reaksi_love' : love,
            'reaksi_wow' : wow})
```
